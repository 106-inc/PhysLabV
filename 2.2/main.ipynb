```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lambdaVal = [
    7032, 6929, 6717, 6678, 6599, 6533, 6507, 6402, 6383, 6334, 6305,
    6267, 6217, 6164, 6143, 6096, 6074, 6030, 5976, 5945, 5882, 5852, 5401,

    5790, 5461, 4395, 4046
]

thetaVal = [
    2942, 2914, 2850, 2830, 2804, 2782, 2770, 2735, 2724, 2708, 2696,
    2682, 2664, 2638, 2638, 2614, 2602, 2584, 2558, 2548, 2516, 2498, 2242,

    2456, 2282, 1866, 1202
]
```

```python
fitCoeffs = np.polyfit(thetaVal, lambdaVal, 2)
poly = np.poly1d(fitCoeffs)
xPoly = np.linspace(0, 3200)
yPoly = poly(xPoly)

plt.figure(figsize=(25, 15))
plt.plot(thetaVal, lambdaVal, 'o')
plt.plot(xPoly, yPoly)

plt.xlim((500, 3500))
plt.ylim((3500, 7500))

plt.rcParams['font.size'] = '28'

plt.xlabel('$\\theta, {}^o$', fontsize=28)
plt.ylabel('$\\lambda, \\AA$', fontsize=28)
plt.grid()
plt.show()
```
